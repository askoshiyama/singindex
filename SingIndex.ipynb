## Fetch every component

```python
df_path = ["C:\\Users\\askos\\Desktop\\SingIndex\\networks\\singindex_networks_avg_extrapolate_2000_2030.csv",
           "C:\\Users\\askos\\Desktop\\SingIndex\\ai\\singindex_ai_avg_extrapolate_2000_2030.csv",
           "C:\\Users\\askos\\Desktop\\SingIndex\\health\\singindex_health_avg_extrapolate_2000_2030.csv",
           "C:\\Users\\askos\\Desktop\\SingIndex\\environment\\singindex_environment_avg_extrapolate_2000_2030.csv",
           "C:\\Users\\askos\\Desktop\\SingIndex\\computing\\singindex_computing_avg_extrapolate_2000_2030.csv",
           "C:\\Users\\askos\\Desktop\\SingIndex\\nanotech\\singindex_nanotech_avg_extrapolate_2000_2030.csv",
           "C:\\Users\\askos\\Desktop\\SingIndex\\social and governance\\singindex_social_avg_extrapolate_2000_2030.csv",
           "C:\\Users\\askos\\Desktop\\SingIndex\\autonomous\\singindex_autonomous_avg_extrapolate_2000_2030.csv",
           "C:\\Users\\askos\\Desktop\\SingIndex\\biotech\\singindex_biotech_avg_extrapolate_2000_2030.csv"]

k=1
for p in df_path:
    if k:
        df = pd.read_csv(p, index_col=0)
        k=0
    else:
        df_new = pd.read_csv(p, index_col=0)
        df = pd.concat([df, df_new], axis=1)
```

```python
df
```

## Aggregate components 

```python
# component weights
component_weights = {"SingIndex Networks": 1.0/9.0,
                     "SingIndex AI": 1.0/9.0,
                     "SingIndex Health": 1.0/9.0,
                     "SingIndex Environment": 1.0/9.0,
                     "SingIndex Computing": 1.0/9.0,
                     "SingIndex Nanotech": 1.0/9.0,
                     "SingIndex Social": 1.0/9.0,
                     "SingIndex Autonomous": 1.0/9.0,
                     "SingIndex Biotech": 1.0/9.0}

# aggregation
singindex = pd.DataFrame({"SingIndex": np.zeros((df.shape[0], ))}, index = df.index)
for c in component_weights.keys():
    singindex["SingIndex"] += df[c].values * component_weights[c]
```

### Visualize

```python
singindex.plot()
```
